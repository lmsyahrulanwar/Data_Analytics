# Exploratory Data Analysis

```python
#import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

#load data
data = pd.read_excel('./testes/Data_SMA.xlsx')
```

* Data yang akan dianalisis adalah data seluruh SMA yang ada di Kota Baubau periode 2022/2023 Ganjil
* Data diporeleh melalui website resmi dari Data Pokok Pendidikan Direktorat Jenderal Pendidikan Anak Usia Dini, Pendidikan Dasar dan Pendidikan Menengah Kementerian Pendidikan, Kebudayaan, Riset dan Teknologi

```python
data.drop(columns='No', axis=1, inplace=True)
print("Ukuran Data : ", data.shape)
data
```

Dari data di atas terdapat 13 data SMA baik negeri maupun swasta yang ada di Kota Baubau

Data pada *data_pegawai.csv* memiliki 14 kolom, yaitu:
* Nama_Sekolah : Nama dari Sekolah
* NPSN : Nomor Pokok Sekolah Nasional
* BP : Bentuk Pendidikan
* Status : Status dari kenggotaan sekolah (swasta/negeri)
* Last_Sync : Sinkronasi data terakhir ke Dapodik
* Jml_Sync : Jumlah data terakhir yang disinkronisasi ke Dapodik
* PD : Jumlah peserta didik
* Rombel : Jumlah rombongan belajar
* Guru : Jumlah guru
* Pegawai : Jumlah pegawai (etc. Tata Usaha, dll)
* R. Kelas : Jumlah ruang kelas
* R. Lab : jumlah ruang laboratorium
* R. Perpus : jumlah ruang perpustakaan

Diketahui data peserta didik terbanyak ada di angka <b>1332</b> siswa dan peserta didik paling sedikit ada di <b>1</b> siswa

```python
sns.set(style="darkgrid")
fig, ax = plt.subplots(figsize=(18,8))
sns.barplot(x='Nama_Sekolah', y='PD', data=data)
plt.xticks(rotation=45, fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Nama Sekolah', fontsize=18)
plt.ylabel('Peserta Didik', fontsize=18)
plt.title('Perbandingan Jumlah Siswa SMA di Kota Baubau', fontsize=25)
plt.tight_layout()
plt.show()
```

```python
sns.set(style="darkgrid")
fig, ax = plt.subplots(figsize=(18,8))
sns.barplot(x='Nama_Sekolah', y='Guru', data=data)
plt.xticks(rotation=45)
plt.xlabel('Nama Sekolah', fontsize=18)
plt.ylabel('Jumlah Guru', fontsize=18)
plt.title('Perbandingan Jumlah Guru SMA di Kota Baubau', fontsize=25)
plt.tight_layout()
plt.show()
```

Dari sini kita dapat memperoleh informasi, yaitu :
* Sekolah yang memiliki guru terbanyak adalah <b>SMAN 2 BAU-BAU</b>
* Sekolah yang memiliki guru paling sedikit adalah <b>SMA PK/LK AL HALIM</b>
* Jumlah guru antara SMAN 1 BAU-BAU dan SMAN 2 BAU-BAU tidak jauh berbeda

Sekarang kita coba visualisasi perbandingan antara sekolah negeri dan swasta di Kota Baubau

```python
negeri = data[data['Status']=='Negeri'].shape[0]
swasta = data[data['Status']=='Swasta'].shape[0]
status = [negeri, swasta]
labels = ['Negeri','Swasta']
# colors = ['orange','blue']

plt.pie(status, labels=labels, autopct='%1.1f%%')
plt.title('Persentasi Sekolah Negeri dan Swasta Di Kota Baubau', fontsize=15)
plt.show()
```

Dari pie chart di atas kita dapat menyimpulkan bahwa sekolah yang berstatus Negeri lebih banyak dibandingkan dengan Swasta tapi dari segi persentase nilainya hampir berimbang dengan selisih <b>7,6%</b> saja
